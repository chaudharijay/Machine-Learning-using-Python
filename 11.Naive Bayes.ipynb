```python
import pandas as pd
```

```python
df= pd.read_csv('11. Naive Bayes titanic.csv')
df
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns',inplace=True)
```

```python
df.head()
```

```python
target = df.Survived
```

```python
inputs = df.drop('Survived', axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex, dtype=int)
dummies.head()
```

```python
inputs = inputs.drop('Sex', axis='columns')
```

```python
inputs
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
inputs
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scaled_input = scale.fit_transform(inputs)
scaled_input
```

```python
new_df = pd.DataFrame(scaled_input,  columns=inputs.columns)
new_df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(new_df, target, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
